Write a program which fill find all such numbers which are divisible by 7 but are not a multiple of 5, between 2000 and 3000 (both included). The numbers obtained should be printed in a comma-seperated sequence on a single line. 

```python
# [x for x in range(2000, 3201) if x % 7 == 0 and x % 5 != 0] 
# - also, make a list for us. Equivalent statement will be: 
# list(x for x in range(2000, 3201) if x % 7 == 0 and x % 5 != 0).

# So [x for x in range(2000, 3201) if x % 7 == 0 and x % 5 != 0] 
# will be like: make us a list of numbers found in the range of 2000 and 3200 
# that are divisible by 7 and not by 5.

list = [x for x in range(2000, 3201) if x % 7 == 0 and x % 5 != 0]

print(list)

# the most confusing is actually the "x for x" expression...
# the first x is for usage and the second is its declaration...
# and this is useful for example...
# imagine that you want to get a list of strings out of this? 
# so we can do something like - ["line - " + str(x) for x in range(2000, 3201) if x % 7 == 0 and x % 5 != 0] 
#this will produce a list with strings like: "line - 2002", "line - 2009", ..., "line - 3199".
```

```python
# The pi value is 3.14 
PI = 3.14

# The  diameter value by default is given in the question
diameter = 12

# The Radius Formula 
radius = diameter/2

# The Formula of Volume
Volume = (4/3) * PI * radius ** 3

# The answer for the volume of a sphere with the given diameter is printed as follows:
print("The Volume of the Sphere is %.2f" %Volume)
```
